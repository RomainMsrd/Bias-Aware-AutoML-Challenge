```python
def plot_group_confusion_matrix(cm, title):
        from matplotlib.patches import Rectangle
        import seaborn as sns

        cm = cm.round(2)
        figsize = (10,8) if len(cm.columns) <= 15 else (20,20)
        fig = plt.figure(figsize=figsize)
        ax = fig.add_subplot()

        sns.heatmap(cm, annot=True, fmt='g', ax=ax, vmin=0, vmax=1, annot_kws={"size": 20});  #annot=True to annotate cells, ftm='g' to disable scientific notation
        
        
        ax.collections[0].colorbar.ax.tick_params(labelsize=16)
        
        # labels, title and ticks
        ax.set_title(title, fontsize=25)
        ax.set_xlabel('Class label Y ', fontsize=20)
        ax.set_ylabel('Spurious feature bin Z', fontsize=20)

        c = cm.to_numpy()
        m = np.amin(c)
        indices = []
        for i in range(3):
            for n in range(3):
                if c[i,n] == m:
                    indices.append((n,i))
        for ind in indices:
            ax.add_patch(Rectangle(ind,1,1, fill=False, edgecolor='blue', lw=3))

        idx = cm.index
        col = cm.columns

        ax.xaxis.set_ticklabels(col, fontsize=16)
        ax.yaxis.set_ticklabels(idx, fontsize=16)
        
        plt.show()
        #super_category_path = os.path.join(PREDICTIONS_PATH, title.replace(" ", "_") + ".png")
        #fig.savefig(super_category_path, dpi=fig.dpi)
```

```python
config = {"Completely_Balanced" : [7,7], "Completely_Biased" : [20,0], "Very_Biased" : [14, 3], "Biased" : [10, 5]}
```

```python
accs = {}
worst_accs = {}
matrices = {}
```

```python
def worst_accuracy_group(cm):
    return np.nanmin(cm.to_numpy())
    
def accuracy(cm):
    return np.nanmean(cm.to_numpy())
```

# Completely Biased/Completely Balanced

```python
cms, datas = [], []

PREDICTIONS_PATH = f"./factsheets/{dataset}_experiment"
TRAIN = COMPLETELY_BIASED
TEST = COMPLETELY_BALANCED
for i in range(100):
    print(i)
    train_conf = config[TRAIN]
    test_conf = config[TEST]
    cm_train, cm_test, data = factsheet_rd(DATASET_PATH, PREDICTIONS_PATH, train_conf, test_conf, SET_SEED = i, GROUP_ACC = True, EPOCHS = 20)
    cms.append(cm_test)
    datas.append(data)

worst_accuracies = [worst_accuracy_group(c) for c in cms]
accuracies = [accuracy(c) for c in cms]

accs[TRAIN + '/' + TEST] = (np.mean(accuracies), np.std(accuracies))
worst_accs[TRAIN + '/' + TEST] = (np.mean(worst_accuracies), np.std(worst_accuracies))
matrices[TRAIN + '/' + TEST] = cms
```

```python
cm = cms[0]*0
for c in cms:
    cm += c
plot_group_confusion_matrix(cm/len(cms), f"{TRAIN + '/' + TEST} Group Accuracy")
```

```python
temp = [m.to_numpy().sum()/3 for m in matrices["biased/biased"]]
accs["biased/biased"] = np.mean(temp), np.std(temp)
```

```python
def get_worst_acc(cms):
    cm = cms[0]*0
    for c in cms:
        cm += c
    cm = cm/len(cms)
    
    ind = None
    c = cm.to_numpy()
    m = np.amin(c)
    for i in range(c.shape[0]):
        for n in range(c.shape[1]):
            if m == c[i,n]:
                ws = [f.to_numpy()[i,n] for f in cms]
                return (np.mean(ws), np.std(ws))
```

```python
def get_worst_acc_diag(cms):
    cm = cms[0]*0
    for c in cms:
        cm += c
    cm = cm/len(cms)
    
    ind = None
    c = cm.to_numpy()
    mini = 100
    t = 0
    for i in range(c.shape[0]):
            if c[i,i] < mini:
                mini = c[i,i]
                t=i
    ws = [f.to_numpy()[t,t] for f in cms]
    return (np.mean(ws), np.std(ws))
```

```python
wst_cc = {k: get_worst_acc(matrices[k]) for k in matrices}
wst_cc["biased/biased"] = get_worst_acc_diag(matrices["biased/biased"])
wst_cc
```
